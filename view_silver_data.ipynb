```python
import os
import duckdb
from dotenv import load_dotenv

load_dotenv()

# That's it! No server, no container needed
con = duckdb.connect()

# Configure to connect to your MinIO
con.execute(f"""
    SET s3_endpoint='localhost:9000';
    SET s3_access_key_id='{os.getenv('MINIO_ROOT_USER')}';
    SET s3_secret_access_key='{os.getenv('MINIO_ROOT_PASSWORD')}';
    SET s3_use_ssl=false;
    SET s3_url_style='path';
""")

# Queries here!
df = con.execute("""
                 
    SELECT * FROM read_parquet('s3://silver/taxi_data.parquet/*.parquet') 
    LIMIT 10;
                 
""").df()

print(f"Number of rows: {len(df)}")
df.head()
```
